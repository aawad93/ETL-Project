```python
#Dependencies
import os
import pandas as pd
```

```python
# List all files in Stocks folder
files = os.listdir("Stocks")
```

```python
# Read all files in Stocks folder into dataframes and append them to list and add Ticker Name column
df_list = []

for f in files:
    try:
        df_list.append(pd.read_csv(f"Stocks/{f}").assign(ticker_name = f.split(sep=".")[0]))
        
    except: 
        print(f)
```

```python
# Combine all my DataFrames
df= pd.concat(df_list)
```
